```python
import pandas as pd
import spacy

nlp = spacy.load("en_core_web_sm")  

df = pd.read_csv("train_reports.csv")
df['Findings_EN'] = df['Findings_EN'].fillna('')  
df['Findings_EN'] = df['Findings_EN'].astype(str) 

caracteristicas = ["effusion", "fibrosis", "nodule", "mass", "atelectasis", "emphysema"]

def extraer_caracteristicas(texto):
    doc = nlp(texto)
    resultado = {}
    for c in caracteristicas:
        resultado[c] = ""
        for sent in doc.sents:
            if c in sent.text.lower():
                resultado[c] = sent.text.strip()
                break
    return resultado


resultado_df = df["Findings_EN"].apply(extraer_caracteristicas).apply(pd.Series)

df_tabular = pd.concat([df, resultado_df], axis=1)
df_tabular.to_csv('tabular_findings')
```
